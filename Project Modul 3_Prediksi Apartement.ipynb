**Analytic Approach**

Jadi, yang dilakukan adalah menganalisis data untuk menentukan harga apartemen yang sesuai dengan mengidentifikasi pola dan hubungan fitur-fitur Apartemen yang tersedia kecuali (Sale Price). Selanjutnya, akan menggunakan algoritma machine larning dengan membuat sebuah **model regresi** yang akan membantu perusahaan untuk menyediakan sebuah **'tools prediction'** harga apartemen Daegu Apartemen.

**Metric Evaluation**

Setelah membuat model regresi, selanjutnya dilakukan dengan mengevaluasi model dengan metrik evaluasi untuk mengukur apakah model yang dihasilkan baik atau tidak. Terdapat 4 metrik evaluasi pada model regresi yaitu MSE, RMSE, MAPE, dan MAE. Pada metrik MSE tidak digunakan karena susahnya untuk interpretasinya.

- **Root Mean Squared Error (RMSE)**

$$RMSE = \sqrt{\frac{1}{n} \sum\limits_{i=1}^n (y_i - \hat{y}_i)^2}$$


- **Mean Absolute Error (MAE)** 
MAE adalah mengukur rata-rata kesalahan absolut antara nilai prediksi dan nilai error. MAE memberikan gambaran seberapa besar kesalahan prediksi rata-rata dalam satuan yang sama dengan data asli. Dengan rumus:

$$MAE = \frac{1}{n} \sum\limits_{i=1}^n |y_i - \hat{y}_i|$$


- **Mean Absolute Percentage Error (MAPE)**
MAPE adalah mengukur presentase rata-rata kesalahan absolut antara nilai prediksi dan nilai error. MAPE memberikan gambaran tentang seberapa besar kesalahan prediksi rata-rata dalam bentuk persentase, yang memudahkan interpretasi terutama ketika membandingkan model pada skala yang berbeda-beda. Dengan rumus:

$$MAPE = \frac{1}{n} \sum\limits_{i=1}^n |\frac{y_i - \hat{y}_i}{y_i}|$$



Dalam evaluasi metrik dapat juga menggunakan R-Squared atau Adj R-Squared, yang digunakan ketika model final menggunakan model linear regresi. Nilai R-Squared adalah ukuran statistik yang menunjukkan seberapa besar variasi suatu variabel dependen dapat dijelaskan oleh variabel independen dalam suatu model regresi.

**Handling Outliers Size_sqf**

```python
sns.histplot(data['Size_sqf'], kde=True)
plt.show()
```

```python
numerical_outliers = ['Size_sqf', 'SalePrice']

def calculate_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

outlier_data = []

for column in numerical_outliers:
    outlier_result, lower_bound, upper_bound = calculate_outliers(data, column)
    total_outlier = len(outlier_result)
    outlier_percentage = round(total_outlier / len(data[column]) * 100, 2)
    outlier_data.append([column, total_outlier, outlier_percentage, lower_bound, upper_bound])

outlier_df = pd.DataFrame(outlier_data, columns=["Column", "Total Outliers", "Percentage (%)", "Lower Bound", "Upper Bound"])
outlier_df
```

- Pada kolom Size(sqf) menghasilkan total outlier sebesar 84, memiliki persentase sebesar 3.11%. Dengan lower bound 107.0 dan upper bound sebesar 1803.0. Pada size(sqf) memiliki jumlah outlier yang cukup tinggi, maka perlu adanya penghapusan outlier
- Pada kolom sale price menghasilkan total outliers sebesar 17, memiliki persentase sebesar 0.63%. Dengan lower bound -67478.5 dan upper bound 521901.5. Untuk kolom Sale Price tidak di bersihkan untuk outlier memingingat Sale Price untuk di analsisi, ketika dihapus akan memperngaruhi analisis dari sale price.

**Encoding**

Encoding merupakan metode yang digunakan untuk mengubah categorical menjadi numerical. Tujuan dari encoding membuat model menjadi mudah untuk di analisis menghasilkan model yang bagus. Pada dataset ini data yang diubah adalah dtaa:
- Kolom TimeToSubway diubah menggunakan tipe Ordinal Encoder, karena memiliki tingkatan
- Kolom HallwayType dan SubwayStation diubah dengan tipe OneHot Encoder

```python
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
from category_encoders import OrdinalEncoder
```

```python
#encoding data ordinal 
ordinal_mapping = [
    {'col':'TimeToSubway','mapping':{'no_bus_stop_nearby':0,'0-5min':1, '5min-10min':2, '10min-15min':3, '15min-20min':4}}
]
```

```python
#Mmebuat model yang akan dianalisis
linreg = LinearRegression()
knn = KNeighborsRegressor()
dtr = DecisionTreeRegressor(random_state=100)
rf = RandomForestRegressor(random_state=100)
xgb = XGBRegressor(random_state=100)
ridge = Ridge(random_state=100)
lasso = Lasso(random_state=100)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_linreg = TransformedTargetRegressor(linreg, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dtr, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)
log_ridge = TransformedTargetRegressor(ridge, func=np.log, inverse_func=np.exp)
log_lasso = TransformedTargetRegressor(lasso, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

models = [log_linreg, log_knn, log_dt, log_rf, log_xgb, log_ridge, log_lasso]

#Membuat calculated untuk R-squared dan Adj R-Squared
def calculate_adjusted_r2(r2, n, p):
    adj_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
    return adj_r2

score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []
adjusted_r_squared = []


score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []
adjusted_r_squared = []

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

model_name = []


for model in models:
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    model_pipeline = Pipeline([
    ('preprocessing', transformer),
    ('scaler', scaler),
    ('model', model)
    ])

    # R-Square
    model_cv_r2 = cross_val_score(
        model_pipeline, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='r2', 
        error_score='raise')
    
    print(model_cv_r2, i)

    score_r2.append(model_cv_r2)
    nilai_mean_r2.append(model_cv_r2.mean())
    nilai_std_r2.append(model_cv_r2.std())

    # Menghitung Adjusted R-squared
    n = len(y_train) 
    p = X_train.shape[1] 
    adj_r2 = calculate_adjusted_r2(model_cv_r2.mean(), n, p)
    adjusted_r_squared.append(adj_r2)

     # RMSE
    model_cv_rmse = cross_val_score(
        model_pipeline, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
        )

    print(model_cv_rmse, model)

    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        model_pipeline, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
        )

    print(model_cv_mae, i)

    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        model_pipeline, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
        )

    print(model_cv_mape, i)

    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())
    model_name.append(models.__class__.__name__)
```

**Hyperparameter Tuning**

Selanjutnya, kita telah menentukan model prediksi test yang lebih baik yaitu **XGBoost Regressor**, maka kita akan menggunakan model **XGBoost Regressor** untuk analisis ini sampai akhir. Pada tahap ini akan dilakukan hyperparameter tuning dengan model XGBoost Regressor, dengan harapan model ini dapat meningkatkan performa yang lebih baik.

```python
max_depth = list(np.arange(1, 11))
learning_rate = list(np.arange(1, 100)/100)
n_estimators = list(np.arange(100, 201))
subsample = list(np.arange(2, 10)/10)
gamma = list(np.arange(1, 11)) 
colsample_bytree = list(np.arange(1, 10)/10)
reg_alpha = list(np.logspace(-3, 1, 10)) 


# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Model XGBoost
xgboost = XGBRegressor(random_state=100)

estimator_xgboost = Pipeline([
    ('preprocessing', transformer),
    ('scaler', scaler),
    ('model', xgboost)
])
```

```python
random_xgboost = RandomizedSearchCV(
    estimator_xgboost,
    param_distributions=hyperparam_space_xgb,
    n_iter=50,
    cv=crossval,
    scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'],
    n_jobs=-1,
    refit='neg_root_mean_squared_error',
    random_state=1
)

random_xgboost.fit(X_train, y_train)
```

```python
pd.DataFrame(random_xgboost.cv_results_).sort_values(by=['rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
print('XGBoost')
print('Best_score:', random_xgboost.best_score_)
print('Best_params:', random_xgboost.best_params_)
```

```python
# Define model terhadap estimator terbaik
xgboost_tuning = random_xgboost.best_estimator_

# Fitting model
xgboost_tuning.fit(X_train, y_train)

# Predict test set
y_pred_xgboost_tuning = xgboost_tuning.predict(X_test)

# Simpan nilai metrics RMSE, MAE, dan MAPE setelah tuning
rmse_xgboost_tuning = np.sqrt(mean_squared_error(y_test, y_pred_xgboost_tuning))
mae_xgboost_tuning = mean_absolute_error(y_test, y_pred_xgboost_tuning)
mape_xgboost_tuning = mean_absolute_percentage_error(y_test, y_pred_xgboost_tuning)

# Hitung R-squared dan adjusted R-squared
r2_xgboost_tuning = r2_score(y_test, y_pred_xgboost_tuning)
n = len(y_test)
p = len(X_test.columns)
adj_r2_xgboost_tuning = 1 - ((1 - r2_xgboost_tuning) * (n - 1) / (n - p - 1))
mape_xgboost_tuning_scaled = mape_xgboost_tuning / 100

# Simpan hasil dalam DataFrame
score_after_tuning_xgboost = pd.DataFrame({
    'R-squared': r2_xgboost_tuning,
    'Adjusted R-squared': adj_r2_xgboost_tuning,
    'RMSE': rmse_xgboost_tuning,
    'MAE': mae_xgboost_tuning,
    'MAPE': mape_xgboost_tuning_scaled  # Gunakan nilai MAPE yang telah diubah ke skala 0 hingga 1
}, index=['XGBoost'])

```
